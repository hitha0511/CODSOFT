```python
ratings.head(5)
```

```python
ratings.shape
```

```python
ratings.rename(columns={
    "User-ID":"User_ID",
    "Book_Rating":"Rating"},inplace = True)
```

```python
ratings.head()
```

```python
ratings['User_ID'].value_counts()
```

```python
x =  ratings['User_ID'].value_counts() > 200
```

```python
ratings=ratings[ratings['User_ID'].isin(y)]
```

```python
ratings
```

```python
ratings.shape
```

```python
books.head(2)
```

```python
ratings_with_books = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={"Book-Rating":"num_of_rating"},inplace=True)
```

```python
num_rating.head()
```

```python
ratings_with_books.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on = 'Title')
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['User_ID','Title'],inplace = True)
```

```python
final_rating.shape
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='User_ID', index='Title', values = 'Book-Rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
books_name = book_pivot.index
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl', 'wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestion)):
        books =  book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = 'A Bend in the Road'
recommend_book(book_name)
```

```python
book_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
recommend_book(book_name)
```

```python
book_name = '4 Blondes'
recommend_book(book_name)
```
